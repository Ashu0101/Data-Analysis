```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([2,9])
y = np.array([1,8])

plt.plot(x,y,marker = "o")
plt.show()
```

```python
x = np.random.normal(150,10,300)
plt.hist(x,color = "red")
plt.xlabel("observation")
plt.ylabel("value")
plt.show()
```

```python
a = np.array([35,25,25,15])
my_labels = np.array(["Apple","Bananas","Cherry","Orange"])
plt.pie(a,labels = my_labels)
plt.show()
```

```python
a = np.array([35,25,25,15])
my_labels = np.array(["Apple","Bananas","Cherry","Orange"])
myexplode =  [0.2,0.8,0.1,0]
plt.pie(a,labels = my_labels,explode = myexplode)
plt.show()
```

```python
a = np.array([35,25,25,15])
my_labels = np.array(["Apple","Bananas","Cherry","Orange"])
plt.pie(a,labels = my_labels)
plt.legend()
plt.title("PIE CHART")
plt.show()
```

```python
plt.bar(my_labels,a,width = 0.5)
plt.show()
```

```python
plt.barh(my_labels,a)
plt.show()
```
