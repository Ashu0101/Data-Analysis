```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Statistics/main/insurance%20(2).csv")
```

```python
df
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(df.age,df.charges,hue = "sex",data = df, palette = ["red","green"],alpha = 0.9)
plt.show()
```

```python
## t-Test to check dependency of gender on charges

Ho = "Charges of men and women are same" ## start of nul hypothesis
Ha = "Charges of men and women are not same" ## start of alternate hypothesis
x = np.array(df[df.sex == "male"].charges) 
y = np.array(df[df.sex == "female"].charges)

t,p_value = stats.ttest_ind(x,y,axis = 0)

if p_value > 0.05:
    print(f'{Ho} as p value ({p_value}) > 0.05')
else:
    print(f"{Ha} as p value ({p_value}) < 0.05")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(df.age,df.bmi,hue = "sex",data = df, palette = ["red","green"],alpha = 0.9)
plt.show()
```

```python
## t-Test to check dependency of gender on bmi

Ho = "BMI of men and women are same" ## start of null hypothesis
Ha = "BMI of men and women are not same" ## start of alternate hypothesis
x = np.array(df[df.sex == "male"].bmi) 
y = np.array(df[df.sex == "female"].bmi)

t,p_value = stats.ttest_ind(x,y,axis = 0)

if p_value > 0.05:
    print(f'{Ho} as p value ({p_value}) > 0.05')
else:
    print(f"{Ha} as p value ({p_value}) < 0.05")
```
